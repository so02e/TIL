```python
import requests
from bs4 import BeautifulSoup
import re

req = requests.get('http://media.daum.net/ranking/popular/') 
html = req.content
html = html.decode('utf-8')
bs = BeautifulSoup(html, 'html.parser')


titles = bs.select('#mArticle > div.rank_news > ul.list_news2 > li > div.cont_thumb > strong > a')
presses = bs.select('#mArticle > div.rank_news > ul.list_news2 > li > div.cont_thumb > strong > span')
news_titles = []
news_presses = [] 



for dom in presses:
    news_presses.append(dom.text)
    
    
for dom in titles:
    news = dom.contents[0]
    news=re.sub(",", ' ', news)
    news_titles.append(news)
    

    
print(news_titles)
print(news_presses)


# 한 행씩 맞춰서 출력
allnewslength = len(news_presses)   
for i in range(allnewslength):
    print("newstitle : " + news_titles[i])
    print("newscomname : " + news_presses[i])

    
# csv 파일로 저장
with open('output/news.csv', "wt", encoding="utf-8") as f:
    f.write('newstitle,newscomname\n')  
    for i in range(len(news_presses)):
        f.write(news_titles[i]+","+news_presses[i]+'\n')  
```

```python
import pandas as pd

data = {'newstitle ' : news_titles ,
        'newscomname ' : news_presses}

df = pd.DataFrame(data)
display(df)

df.to_csv("./output/news_new.csv")
```
