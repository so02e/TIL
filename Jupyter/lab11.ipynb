```python
import pandas as pd

file_path = './data/score.csv'

df1 = pd.read_csv(file_path)

print(df1.columns)
print(df1.index)
df1['total'] = df1.kor + df1.eng + df1.mat
df1['avg'] = (df1.kor + df1.eng + df1.mat) / 3
pd.options.display.float_format = '{:.2f}'.format

display(df1)
```

```python
file_path = './data/score_noheader.csv'

df2 = pd.read_csv(file_path, header=None)

display(df2)

print(df2.columns)
print(df2.index)


print("---------컬럼명 추가-----------")
df2.columns=['name', 'kor', 'eng', 'mat']
display(df2)
print(df2.columns)

print("---------total과 avg 열 추가-----------")
df2['total'] = df2.kor + df2.eng + df2.mat
df2['avg'] = (df2.kor + df2.eng + df2.mat) / 3
display(df2)
```

```python
file_path = './data/score_header.csv'

df3 = pd.read_csv(file_path, skiprows=3)

display(df3)

print(df3.columns)
print(df3.index)

df3['total'] = df3.kor + df3.eng + df3.mat
df3['avg'] = (df3.kor + df3.eng + df3.mat) / 3
display(df3)


# 헤더와 인덱스는 출력하지 않고, 값을 세번씩 반복하여 csv 파일에 저장되어야 함. 
df3.to_csv("./output/score_result.csv", index=None, header = None ) 
df3.to_csv("./output/score_result.csv", mode = 'a', index=None, header = None ) 
df3.to_csv("./output/score_result.csv", mode = 'a', index=None, header = None ) 
```

```python
df4 = pd.read_excel('./data/score.xlsx')         

display(df4)

df4['total'] = df4.kor + df4.eng + df4.mat
df4['avg'] = (df4.kor + df4.eng + df4.mat) / 3
df4 = df4.sort_values(by='total',ascending=False)

display(df4)

# 헤더 출력, 인덱스 출력
writer1 = pd.ExcelWriter("./output/score_result1.xlsx")
df4.to_excel(writer1, sheet_name="sheet1")
writer1.save()

# 헤더 출력, 인덱스 미출력
writer2 = pd.ExcelWriter("./output/score_result2.xlsx")
df4.to_excel(writer2, sheet_name="sheet1", index=None)
writer2.save()
```
