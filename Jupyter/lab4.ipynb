```python
import re
import requests
from bs4 import BeautifulSoup

toon_title=[]
toon_summary=[]
toon_rating=[]

for n in range(1,11):
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?page='+str(n))
    html = req.content
    html = html.decode('utf-8')
    bs = BeautifulSoup(html, 'html.parser')

    
    title = bs.select('div.challengeInfo > h6 > a')
    summary = bs.select('div.challengeInfo > div.summary')
    rating = bs.select('div.challengeInfo > div.rating_type > strong')


    for titledom in title:
        toon_title.append(titledom.text.strip())
    for summarydom in summary:
        toon_summary.append(summarydom.text.strip())
    for ratingdom in rating:
        toon_rating.append(ratingdom.text.strip())

print(toon_title)
print(toon_summary)
print(toon_rating)
```

```python
import requests
from bs4 import BeautifulSoup
import re


for n in range(1,11):
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?page='+str(n)) # 페이지 이동. 
    html = req.content
    html = html.decode('utf-8')
    bs = BeautifulSoup(html, 'html.parser')

    titles = bs.select('div.challengeInfo > h6 > a')
    summarys = bs.select('div.challengeInfo > div.summary')
    ratings = bs.select('div.challengeInfo > div.rating_type > strong')
    
    comic_title = []
    comic_summary = [] 
    comic_rating = []


    for title in titles:
        content = title.contents[0]
        content = re.sub("[\n\t]", '', content)
        comic_title.append(content.strip())
    print(comic_title)
    
    
    
    for summary in summarys:
        comic_summary.append(summary.text)
    print(comic_summary)


                
    for rating in ratings:
        comic_rating.append(rating.text)
    print(comic_rating)

```
